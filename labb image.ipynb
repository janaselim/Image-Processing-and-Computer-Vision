```python
import numpy as np
import cv2
from PIL import Image
import math
import matplotlib.pyplot as plt
```

```python
person = cv2.imread("/content/student.jpg")
person=cv2.cvtColor(person,cv2.COLOR_BGR2RGB)
background = cv2.imread("/content/giu.jpg")
background=cv2.cvtColor(background,cv2.COLOR_BGR2RGB)
# Create a figure and subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 6))
# Display the gray image
axs[0].imshow(person, cmap='gray')
axs[0].set_title('person Image')
# Display the gray image
axs[1].imshow(background, cmap='gray')
axs[1].set_title('background Image')
plt.show()
```

```python
person = cv2.resize(person, (background.shape[1], background.shape[0]))
# Blend the images
result = cv2.add(person,background)
# Display the result
plt.imshow(result) # Convert BGR to RGB for Matplotlib
```

```python
# Blend the images
result2 = cv2.addWeighted(person, 0.5, background, 0.5, 0)
plt.imshow(result2)
plt.show()
```

```python
person=cv2.cvtColor(person,cv2.COLOR_BGR2RGB)
gray_img=cv2.cvtColor(person,cv2.COLOR_RGB2GRAY)
plt.imshow(gray_img,cmap='gray')
```

```python
threshold_value = 250
# Apply binary thresholding
_, mask = cv2.threshold(gray_img, threshold_value, 255,
cv2.THRESH_BINARY)
plt.imshow(mask,cmap='gray')
```

```python
mask_inv = cv2.bitwise_not(mask)
plt.imshow(mask_inv,cmap='gray')
```

```python
person = cv2.bitwise_and(person, person, mask=mask_inv)
person=cv2.cvtColor(person,cv2.COLOR_BGR2RGB)
plt.imshow(person)
```

```python
background = cv2.bitwise_and(background, background, mask=mask)
plt.imshow(background)
```

```python
result = cv2.add(background, person)
plt.imshow(result)
```
