```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
image = cv2.imread('img.jpeg')

red_channel, green_channel, blue_channel = cv2.split(image)
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(blue_channel, cv2.COLOR_BGR2RGB))
plt.title('Blue Channel')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(green_channel, cv2.COLOR_BGR2RGB))
plt.title('Green Channel')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(red_channel, cv2.COLOR_BGR2RGB))
plt.title('Red Channel')

```

```python
merged = cv2.merge([blue_channel, green_channel, red_channel])
plt.figure(figsize=(10, 5))
plt.imshow(cv2.cvtColor(merged, cv2.COLOR_BGR2RGB))
plt.title('Merged Image')
```

```python
blue_channel, green_channel, red_channel = cv2.split(image)

cyan_channel = 255 - red_channel
magenta_channel = 255 - green_channel
yellow_channel = 255 - blue_channel
```

```python
from google.colab.patches import cv2_imshow
```

```python
cmyk_image = cv2.merge((cyan_channel, magenta_channel, yellow_channel))
cv2_imshow('cmyk_image')
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

noisy_img = cv2.imread('salt.jpeg', 0)

k1 = np.ones((3, 3)) / 9.0

k2 = np.array([
              [1/16, 2/16, 1/16],
              [2/16, 4/16, 2/16],
              [1/16, 2/16, 1/16]
              ])

filtered_img1 = cv2.filter2D(noisy_img, -1, k1)
filtered_img2 = cv2.filter2D(noisy_img, -1, k2)

plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.imshow(noisy_img, cmap='gray')
plt.title('Noisy Image')

plt.subplot(1, 3, 2)
plt.imshow(filtered_img1, cmap='gray')
plt.title('Average Filtered Image')

plt.subplot(1, 3, 3)
plt.imshow(filtered_img2, cmap='gray')
plt.title('Weighted Filtered Image')

plt.show()
```
