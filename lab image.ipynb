```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
salt_image=cv2.imread("saltandpepper.png")
plt.imshow(salt_image,cmap="gray")
```

```python
filtered_img=cv2.medianBlur(salt_image,5)
plt.imshow(filtered_img,cmap="gray")
```

```python
#speckle b mean
mri=cv2.imread("mri.jpg")
plt.imshow(mri,cmap="gray")
```

```python
kernel_size=11
kernel=np.ones((kernel_size,kernel_size))/(kernel_size*kernel_size)
filtered_image=cv2.filter2D(mri,-1,kernel)
plt.imshow(filtered_image,cmap="gray")
```

```python
#sharpening
img=cv2.imread("All_Gizah_Pyramids.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
sharpening_kernel=np.array([
    [-1,-1,-1],
    [-1,9,-1],
    [-1,-1,-1]
    ])
sharpened_img=cv2.filter2D(img,-1,sharpening_kernel)
plt.imshow(sharpened_img)
```

```python
#edge detection
image=cv2.imread("streetsigns.jpg",0)#0 ashan gray alatoool
sobel_x_kernel=np.array([
                        [-1,0,1],
                        [-2,0,2],
                        [-1,0,1]
                        ])
sobel_y_kernel=np.array([
                        [-1,-2,-1],
                        [0,0,0],
                        [1,2,1]
                        ])
image_sobel_x=cv2.filter2D(image,-1,sobel_x_kernel)
image_sobel_y=cv2.filter2D(image,-1,sobel_y_kernel)
plt.imshow(image_sobel_x,cmap='gray')
plt.imshow(image_sobel_y,cmap='gray')
```
